O que é clustering? Seus principais algoritmos? Sua aplicação?

R. Técnica de aprendizado não supervisionado que visa agrupar dados em clusters ou grupos, onde os elementos dentro de um mesmo grupo são mais semelhantes entre si do que em relação a elementos de outros grupos. Essa técnica é amplamente utilizada para explorar e entender grandes volumes de dados.

**K-means** é um dos algoritmos de clustering mais populares. Ele funciona dividindo um conjunto de dados em *K grupos*, onde cada grupo é representado por um centróide. O algoritmo iterativamente atribui cada ponto ao grupo cujo centróide está mais próximo, recalculando os centróides até a convergência.

**Hierarchical Clustering** hierarquia de clusters, que pode ser visualizada como uma árvore (dendrograma). Existem duas abordagens: aglomerativa (começa com cada ponto como um cluster e une-os) e divisiva (começa com um único cluster e o divide).

**DBSCAN (Density-Based Spatial Clustering of Applications with Noise)** algoritmo baseado em densidade que agrupa pontos próximos uns dos outros e identifica pontos isolados como ruído. Ele requer dois parâmetros: a distância máxima (epsilon) e o número mínimo de pontos para formar um cluster.

**Agrupamento de Dados sem Rótulos** não se possui rótulos para os dados, clustering é útil para descobrir padrões ou estruturas ocultas. Isso é comum em análise exploratória de dados. A interpretação dos clusters pode ser subjetiva, e a qualidade do agrupamento pode variar dependendo do algoritmo e dos parâmetros escolhidos. Além disso, dados ruidosos podem levar a resultados menos confiáveis.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# Exemplo de dados (substitua isso pelo seu DataFrame)
data = {
    'Smoking_Prevalence': np.random.rand(100) * 100,
    'Drug_Experimentation': np.random.rand(100) * 100,
}
df = pd.DataFrame(data)

# Escalonar os dados
scaler = StandardScaler()
features_scaled = scaler.fit_transform(df)

# Definir o número de clusters
k = 10  # Você pode ajustar esse valor
kmeans = KMeans(n_clusters=k, random_state=42)
kmeans.fit(features_scaled)

# Adicionar os rótulos dos clusters ao DataFrame
df['Cluster'] = kmeans.labels_

# Visualizar os clusters
plt.figure(figsize=(10, 6))
plt.scatter(df['Smoking_Prevalence'], df['Drug_Experimentation'], c=df['Cluster'], cmap='viridis', alpha=0.6)
plt.title('Clustering de Jovens em Função da Prevalência de Tabagismo e Experimentação com Drogas')
plt.xlabel('Prevalência de Tabagismo')
plt.ylabel('Experimentação com Drogas')
plt.colorbar(label='Cluster')
plt.show()
```

Eixo X (Prevalência de Tabagismo): Representa a porcentagem de jovens que relataram fumar.
Se um ponto está na posição "30" no eixo X, isso significa que 30% dos jovens desse grupo específico afirmaram ser usuários de tabaco.

Eixo Y (Experimentação com Drogas): Representa a porcentagem de jovens que relataram ter experimentado drogas. Um ponto na posição "50" no eixo Y indica que 50% dos jovens do grupo afirmaram ter experimentado drogas.

**Os números não são simbólicos;** eles representam dados quantitativos obtidos do dataset. Eles refletem a prevalência real e a experimentação entre os jovens.

Cada ponto no gráfico representa um registro individual do dataset, ou seja, um grupo de jovens com características específicas em relação ao tabagismo e à experimentação com drogas.

Os pontos são coloridos com base nos clusters identificados pelo K-means. Cada cor representa um grupo distinto de jovens que compartilham características semelhantes em relação à prevalência de tabagismo e experimentação com drogas.

Cores diferentes indicam grupos que podem ter comportamentos ou características diferentes.

Os números em si (como 0, 1, 2) não têm um significado específico em termos de valores de dados, mas sim representam a identificação dos clusters.
Os pontos coloridos (baseados nos clusters) mostram como os jovens estão agrupados de acordo com a prevalência de tabagismo e a experimentação com drogas. Cada cor reflete um grupo de comportamento similar.

Se muitos pontos de um determinado cluster estão concentrados em uma área específica do gráfico, isso sugere que esse grupo tem características comuns em relação à prevalência de tabagismo e experimentação com drogas.

Os jovens nesse cluster (amarelo) podem apresentar padrões de comportamento similares. Por exemplo:
Alta Prevalência de Tabagismo: Se a maioria dos pontos amarelos estiver em uma área mais alta do eixo X, isso indica que esses jovens têm uma alta taxa de tabagismo.
Alta Experimentação com Drogas: Se os pontos estiverem mais para cima no eixo Y, isso sugere que eles também têm uma alta taxa de experimentação com drogas.

PCA (Análise de Componentes Principais) para reduzir a dimensionalidade dos dados.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

# Exemplo de dados (substitua isso pelo seu DataFrame)
data = {
    'Smoking_Prevalence': np.random.rand(100) * 100,
    'Drug_Experimentation': np.random.rand(100) * 100,
    'Other_Feature': np.random.rand(100) * 100  # Exemplo de mais uma dimensão
}
df = pd.DataFrame(data)

# Escalonar os dados
scaler = StandardScaler()
features_scaled = scaler.fit_transform(df)

# Definir o número de clusters
k = 3  # Você pode ajustar esse valor
kmeans = KMeans(n_clusters=k, random_state=42)
kmeans.fit(features_scaled)

# Adicionar os rótulos dos clusters ao DataFrame
df['Cluster'] = kmeans.labels_

# Aplicar PCA para reduzir a dimensionalidade
pca = PCA(n_components=2)
principal_components = pca.fit_transform(features_scaled)

# Criar um novo DataFrame com os componentes principais
df_pca = pd.DataFrame(data=principal_components, columns=['PC1', 'PC2'])
df_pca['Cluster'] = df['Cluster']

# Visualizar os clusters
plt.figure(figsize=(10, 6))
plt.scatter(df_pca['PC1'], df_pca['PC2'], c=df_pca['Cluster'], cmap='viridis', alpha=0.6)
plt.title('Clusters gerados pelo K-Means após PCA')
plt.xlabel('Componente Principal 1')
plt.ylabel('Componente Principal 2')
plt.colorbar(label='Cluster')
plt.show()
```

Reduzir a dimensionalidade dos dados enquanto preserva o máximo de variação possível

PCA transforma os dados originais em um novo sistema de coordenadas, onde os eixos (componentes principais) são ordenados de acordo com a quantidade de variância que explicam.

A forma dos clusters pode fornecer insights sobre a natureza dos dados. Por exemplo, clusters mais compactos podem indicar grupos homogêneos, enquanto clusters mais dispersos podem sugerir diversidade dentro do grupo.

**Clusters Identificados** no gráfico que mostra a prevalência de tabagismo em relação à experimentação com drogas, os pontos estão agrupados em diferentes clusters, indicados por cores distintas.
Cluster 1: Pode ser visualizado na parte superior do gráfico, indicando alta prevalência de tabagismo e variabilidade na experimentação com drogas.
Cluster 2: Localizado em uma região onde a prevalência de tabagismo é baixa, mas a experimentação com drogas pode ser mais alta, sugerindo um perfil diferente.
Cluster 3: Os pontos podem estar distribuídos em uma faixa intermediária, refletindo características mistas.
Características de Cada Cluster
Cluster 1:

Descrição: Este grupo pode representar indivíduos que têm hábitos mais arraigados de tabagismo, mas que não necessariamente se envolvem em experimentação de drogas.
Interpretação: Esses indivíduos podem estar mais focados em hábitos relacionados ao tabagismo e menos propensos a experimentar drogas.

Cluster 2:

Descrição: Este cluster apresenta indivíduos com baixa prevalência de tabagismo, mas que se envolvem mais frequentemente em experimentação com drogas.
Interpretação: Esses jovens podem ser mais exploratórios em suas experiências, sugerindo uma predisposição a experimentar substâncias, mas não se identificando com o tabagismo.

Cluster 3:

Descrição: Representa indivíduos com características intermediárias entre os outros dois grupos.
Interpretação: Esse grupo pode incluir aqueles que estão em transição ou que têm hábitos variados, mostrando que podem ser influenciados por fatores sociais ou ambientais.

**Perfis de Clientes e Insights**  Cluster 1 (Conservadores):

Perfil: Indivíduos que podem estar cientes dos riscos associados ao tabagismo, mas continuam com esse hábito.
Insight: Estratégias de marketing podem focar em campanhas de conscientização sobre os efeitos nocivos do tabagismo e promover recursos para ajudar na cessação.

Cluster 2 (Exploradores):

Perfil: Jovens mais abertos a experiências novas, mas que não se comprometem com o tabagismo.
Insight: Programas educativos sobre os riscos da experimentação com drogas poderiam ser mais impactantes neste grupo, abordando as consequências potenciais de tais comportamentos.

Cluster 3 (Moderados):

Perfil: Indivíduos que têm um comportamento misto em relação ao tabagismo e uso de drogas.
Insight: Intervenções que abordem tanto o tabagismo quanto a experimentação com drogas podem ser eficazes, dado que esse grupo pode estar mais suscetível a influências externas.

Com um número menor de clusters, a segmentação tende a ser mais ampla, o que pode resultar em grupos mais heterogêneos. Isso pode dificultar a identificação de perfis específicos de clientes, pois diferentes comportamentos podem ser agrupados juntos.

Número maior de clusters permite uma segmentação mais refinada, com grupos que representam características mais específicas. Isso pode ajudar a identificar perfis distintos, como jovens com altas taxas de tabagismo versus aqueles com baixa experimentação de drogas.

```python
import pandas as pd

# Ajuste o nome do arquivo conforme necessário
df = pd.read_csv('dataset/car_price_prediction_.csv')  # Substitua pelo nome correto do arquivo

# Exibir as primeiras linhas do DataFrame
print(df.head())
```

Características para Clustering

```python
# Selecionar características relevantes
features = df[['Year', 'Mileage', 'Price']]

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
features_scaled = scaler.fit_transform(features)
```

**DBSCAN**

```python
from sklearn.cluster import DBSCAN
import matplotlib.pyplot as plt

# Aplicar DBSCAN
dbscan = DBSCAN(eps=0.5, min_samples=5)
clusters_dbscan = dbscan.fit_predict(features_scaled)

# Adicionar os rótulos dos clusters ao DataFrame
df['DBSCAN_Cluster'] = clusters_dbscan

# Visualizar os clusters
plt.figure(figsize=(10, 6))
plt.scatter(df['Year'], df['Price'], c=df['DBSCAN_Cluster'], cmap='viridis', alpha=0.6)
plt.title('DBSCAN Clustering')
plt.xlabel('Year')
plt.ylabel('Price')
plt.colorbar(label='Cluster')
plt.show()
```

**Agglomerative Clustering**

```python
from sklearn.cluster import AgglomerativeClustering

# Aplicar Agglomerative Clustering
agglo = AgglomerativeClustering(n_clusters=3)  # Você pode ajustar o número de clusters
clusters_agglo = agglo.fit_predict(features_scaled)

# Adicionar os rótulos dos clusters ao DataFrame
df['Agglo_Cluster'] = clusters_agglo

# Visualizar os clusters
plt.figure(figsize=(10, 6))
plt.scatter(df['Year'], df['Price'], c=df['Agglo_Cluster'], cmap='viridis', alpha=0.6)
plt.title('Agglomerative Clustering')
plt.xlabel('Year')
plt.ylabel('Price')
plt.colorbar(label='Cluster')
plt.show()
```

**DBSCAN** é eficaz para identificar outliers, pois rotula pontos que não pertencem a nenhum cluster como ruído.

DBSCAN forma clusters com base na densidade de pontos e é capaz de identificar clusters de formas arbitrárias, além de rotular outliers como ruído. Já Agglomerative Clustering cria uma hierarquia de clusters, unindo iterativamente os mais próximos, e é mais adequado para dados que seguem uma estrutura hierárquica.

A escolha dos parâmetros eps (raio de proximidade) e min_samples (número mínimo de pontos para formar um cluster) é crucial. Um eps pequeno pode resultar em muitos clusters pequenos e ruído, enquanto um eps grande pode levar à formação de poucos clusters grandes, potencialmente ignorando a estrutura local dos dados.

Características como Year, Mileage e Price são frequentemente relevantes. Ao analisar os clusters formados, podemos verificar se, por exemplo, os carros mais novos tendem a ser agrupados juntos e como a quilometragem influencia o preço.

Podemos usar métricas como o Silhouette Score, que mede a similaridade dos pontos dentro de um cluster em relação a pontos em clusters diferentes. Um Silhouette Score próximo de +1 indica que os pontos estão bem agrupados. O Dunn Index também pode ser utilizado para avaliar a separação entre clusters.

A análise dos clusters pode revelar tendências no mercado, como a existência de um segmento de carros de luxo com alta quilometragem, ou a concentração de carros mais novos a preços acessíveis. Isso ajuda a entender melhor o comportamento do consumidor e as dinâmicas de preços.
